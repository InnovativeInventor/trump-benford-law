```python
import json
import seaborn as sns
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
## Benford's Law
x = np.array(range(1,10))
y = np.log10(1+(1)/x)
```

```python
## Trump's tweets
numbers = []
for each_year in range(2015,2018):
    with open("condensed_" + str(each_year) + ".json",'r') as json_file:
        all_tweets = json.load(json_file)
        for each_tweet in all_tweets:
            tweet_text = each_tweet['text']
            if [int(s[0]) for s in tweet_text.split() if s.isdigit()]: # Don't you wish that Python 3.8 was released?
                numbers.extend([int(s[0]) for s in tweet_text.split() if s.isdigit() and len(s) != 4])
```

```python
print(numbers)
```

```python
plt.xlim(1, 9)
sns.distplot(numbers, label="Trump's Tweets",bins=10)
plt.plot(x,y,label="Benford's Law")
plt.legend()
plt.title("Trump's Tweets Compared to Benford's Law")
plt.xlabel("First digit")
plt.ylabel("% occurance")
plt.savefig("trump-benford-law.png")
```
